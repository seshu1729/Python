### Step 7

```python
from pptx import Presentation
from pptx.util import Inches, Pt

def add_content_to_slide(content_box, bullets):
    tf = content_box.text_frame

    for bullet in bullets:
        p = tf.add_paragraph()  # New paragraph for each bullet

        if "**" in bullet:  # Handle bold formatting
            parts = bullet.split("**")
            for i, part in enumerate(parts):
                if part:  # Skip empty parts
                    run = p.add_run()
                    run.text = part
                    run.font.size = Pt(18)
                    run.font.bold = (i % 2 == 1)
        else:
            p.text = bullet
            p.font.size = Pt(18)

# Test it
prs = Presentation()
slide_layout = prs.slide_layouts[6]
slide = prs.slides.add_slide(slide_layout)

# Add textbox
content_box = slide.shapes.add_textbox(Inches(1), Inches(1), Inches(8), Inches(5))

# Example bullets
bullets = [
    "Python is **easy**",
    "**Automation** saves time",
    "We can mix **normal** and bold"
]

add_content_to_slide(content_box, bullets)

prs.save("bullets_bold.pptx")
print("Bullets + bold slide created!")
```

### Step 8

```python
from pptx import Presentation
from pptx.util import Inches, Pt

MAX_BULLETS = 6

def add_content_to_slide(content_box, bullets):
    tf = content_box.text_frame
    for bullet in bullets:
        p = tf.add_paragraph()
        p.text = bullet
        p.font.size = Pt(18)

prs = Presentation()
slide_layout = prs.slide_layouts[6]

# Section title
section_title = "Demo Section"

# First slide
slide = prs.slides.add_slide(slide_layout)
title_box = slide.shapes.add_textbox(Inches(0.5), Inches(0.5), Inches(8), Inches(1))
title_box.text_frame.text = section_title
content_box = slide.shapes.add_textbox(Inches(0.5), Inches(1.5), Inches(8), Inches(5))

# Example bullets (more than 6)
all_bullets = [f"Point {i}" for i in range(1, 15)]

current_bullets = []
for bullet in all_bullets:
    current_bullets.append(bullet)

    if len(current_bullets) == MAX_BULLETS:
        add_content_to_slide(content_box, current_bullets)
        current_bullets = []

        # New continuation slide
        slide = prs.slides.add_slide(slide_layout)
        title_box = slide.shapes.add_textbox(Inches(0.5), Inches(0.5), Inches(8), Inches(1))
        title_box.text_frame.text = section_title + " (contd.)"
        content_box = slide.shapes.add_textbox(Inches(0.5), Inches(1.5), Inches(8), Inches(5))

# Handle leftover bullets
if current_bullets:
    add_content_to_slide(content_box, current_bullets)

prs.save("split_slides.pptx")
print("Slides with auto-splitting created!")
```
